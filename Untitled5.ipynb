```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from google.colab import files
import io

sns.set(style="whitegrid")  # Mejora estilo gráfico
```

Paso 2: Carga y exploración del dataset
Explicación:
Cargamos el archivo CSV desde el computador local y exploramos sus características principales: dimensiones, columnas, valores faltantes y un vistazo a las primeras filas. Esto nos permite conocer la estructura y calidad de los datos para tomar decisiones.

```python
uploaded = files.upload()
filename = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[filename]), sep=';')

print(f"Dimensiones del dataset: {df.shape}")
print("Columnas:")
print(df.columns.tolist())

print("\nPrimeras filas:")
display(df.head())

print("\nValores nulos por columna:")
print(df.isnull().sum())

print("\nEstadísticas descriptivas del monto préstamo digital:")
display(df['ventaPrestDig'].describe())
```

Paso 3: Regresión lineal simple
Explicación:
Seleccionamos la variable independiente trxDigitalUm (número de transacciones digitales) porque refleja la actividad financiera digital del cliente, que se espera influya en el monto de préstamo otorgado. Ajustamos un modelo de regresión lineal simple para cuantificar esta relación.

```python
# Preparar datos para regresión lineal simple
df_rls = df.dropna(subset=['trxDigitalUm', 'ventaPrestDig'])
df_rls['trxDigitalUm'] = pd.to_numeric(df_rls['trxDigitalUm'], errors='coerce')
df_rls['ventaPrestDig'] = pd.to_numeric(df_rls['ventaPrestDig'], errors='coerce')
df_rls = df_rls.dropna(subset=['trxDigitalUm', 'ventaPrestDig'])

X = df_rls[['trxDigitalUm']].values
y = df_rls['ventaPrestDig'].values

# Entrenar modelo
modelo_rls = LinearRegression()
modelo_rls.fit(X, y)

# Predecir valores
y_pred = modelo_rls.predict(X)

# Calcular métricas
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("Métricas del modelo de regresión lineal simple:")
print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"R2: {r2:.4f}")
```

Explicación:
El MAE indica el error medio en unidades monetarias, el MSE penaliza errores grandes, y el R2 indica qué tanto varía el monto explicado por la variable independiente. Valores mejores significan un modelo más confiable.

```python
# Visualización del modelo
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3, label='Datos reales')
plt.plot(X, y_pred, color='red', label='Predicción Regresión Lineal')
plt.xlabel('Número de transacciones digitales')
plt.ylabel('Monto préstamo digital')
plt.title('Regresión Lineal Simple')
plt.legend()
plt.show()
```

```python
# Análisis de residuos
residuos = y - y_pred
plt.figure(figsize=(10,6))
plt.scatter(y_pred, residuos, alpha=0.3)
plt.hlines(0, xmin=y_pred.min(), xmax=y_pred.max(), colors='red')
plt.xlabel('Valores predichos')
plt.ylabel('Residuos')
plt.title('Análisis de residuos')
plt.show()
```

Paso 4: Red neuronal con múltiples variables
Explicación:
Para mejorar la predicción, usamos una red neuronal que toma múltiples variables, incluyendo variables categóricas codificadas (como tipo de operación digital y procedencia) y la variable numérica trxDigitalUm. Esto permite capturar relaciones más complejas entre características y el monto de préstamo.

```python
# Codificar variables categóricas
cat_vars = ['operDigital', 'procedencia']
for var in cat_vars:
    df[var] = df[var].astype(str)  # asegurar tipo string

le_dict = {}
for var in cat_vars:
    le = LabelEncoder()
    df[var + '_cod'] = le.fit_transform(df[var])
    le_dict[var] = le

# Preparar variables para red neuronal
features = ['operDigital_cod', 'procedencia_cod', 'trxDigitalUm']
df_nn = df.dropna(subset=features + ['ventaPrestDig'])
X_nn = df_nn[features].values
y_nn = df_nn['ventaPrestDig'].values

# Escalar datos
scaler = StandardScaler()
X_nn_scaled = scaler.fit_transform(X_nn)

# Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_nn_scaled, y_nn, test_size=0.2, random_state=42)
```

```python
# Construcción del modelo
model_nn = Sequential([
    Dense(32, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(16, activation='relu'),
    Dense(1)
])

model_nn.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Entrenamiento
history = model_nn.fit(X_train, y_train, epochs=60, batch_size=64, validation_split=0.2, verbose=1)
```

Explicación:
La red neuronal tiene dos capas ocultas con función ReLU para capturar relaciones no lineales. Se usa MSE como función de pérdida y MAE como métrica para evaluar precisión durante entrenamiento.

```python
# Graficar evolución del error y MAE
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='Pérdida entrenamiento')
plt.plot(history.history['val_loss'], label='Pérdida validación')
plt.xlabel('Épocas')
plt.ylabel('MSE')
plt.legend()
plt.title('Evolución de pérdida')

plt.subplot(1,2,2)
plt.plot(history.history['mae'], label='MAE entrenamiento')
plt.plot(history.history['val_mae'], label='MAE validación')
plt.xlabel('Épocas')
plt.ylabel('MAE')
plt.legend()
plt.title('Evolución de MAE')

plt.tight_layout()
plt.show()
```

```python
# Evaluación en test
y_pred_nn = model_nn.predict(X_test).flatten()
mae_nn = mean_absolute_error(y_test, y_pred_nn)
mse_nn = mean_squared_error(y_test, y_pred_nn)
r2_nn = r2_score(y_test, y_pred_nn)

print(f"Red neuronal - MAE: {mae_nn:.2f}, MSE: {mse_nn:.2f}, R2: {r2_nn:.4f}")
```

```python
# Comparación valores reales vs predichos
plt.figure(figsize=(10,6))
sns.scatterplot(x=y_test, y=y_pred_nn, alpha=0.3)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel('Monto real')
plt.ylabel('Monto predicho')
plt.title('Valores reales vs predichos - Red neuronal')
plt.show()

# Residual plot red neuronal
residuals_nn = y_test - y_pred_nn
plt.figure(figsize=(10,6))
sns.scatterplot(x=y_pred_nn, y=residuals_nn, alpha=0.3)
plt.hlines(0, xmin=y_pred_nn.min(), xmax=y_pred_nn.max(), colors='red')
plt.xlabel('Valores predichos')
plt.ylabel('Residuos')
plt.title('Análisis de residuos - Red neuronal')
plt.show()
```

Paso 5: Uso de estructuras de programación (bucle, condición, lista, diccionario)
Explicación:
Se utiliza un bucle para contar la cantidad de registros por tipo de operación digital, guardando el conteo en un diccionario. La lista de tipos se obtiene con una función de pandas.

```python
operaciones = df['operDigital'].unique().tolist()
conteo_oper = {}

for op in operaciones:
    if op not in conteo_oper:
        conteo_oper[op] = 0
    conteo_oper[op] += len(df[df['operDigital'] == op])

print("Conteo de registros por tipo de operación digital:")
for k,v in conteo_oper.items():
    print(f"{k}: {v}")

plt.figure(figsize=(12,6))
sns.barplot(x=list(conteo_oper.keys()), y=list(conteo_oper.values()))
plt.xticks(rotation=45)
plt.title('Cantidad de registros por tipo de operación digital')
plt.xlabel('Tipo de operación')
plt.ylabel('Cantidad')
plt.show()
```
